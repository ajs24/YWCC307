### **Runtime analysis of password guessing algorithm**

This is a simple bruteforce algorithm that guesses every possible password up to length 7 that contains ***only lowercase characters***. When running this script, the user is prompted for input. The length of that input is determined using the len function. Nested loops are used to exhaustively search every possible string of that length until the password is correctly guessed.

* There are **$26$** possible passwords of length $1$, the characters a - z.
* There are **$676$** possible passwords of length $2$. aa, ab, ac, ..., zz
* There are **$17,576$** possible passwords of length $3$. aaa, aab, aac, ..., zzz.
* There are **$456,976$** possible passwords of length $4$.
* There are **$11,881,376$** possible passwords of length $5$.
* There are **$308,915,776$** possible passwords of length $6$.
* There are **$8,031,810,176$** possible passwords of length $7$.
* There are $26^{n}$ possible passwords of length $n$.


Below this code is the corresponding input used and output generated for each function call. We can see that as the length of our password increases, the time it takes to crack that password via brute force increases exponentially. Guessing "gophers" took more than 2 billion attempts!


```python
import time

def brute():
	x=input("Enter a word, all lower case.\n")
	lca1='abcdefghijklmnopqrstuvwxyz'

	answer = ''
	possibilities = 0

	if(len(x) == 1):
		for letter1 in lca1:
			possibilities += 1
			answer = letter1
			if x == answer:
				print("The answer is " + answer + ". It took " + "{:,}".format(possibilities) + " guesses.")
				return

	if(len(x) == 2):
		for letter1 in lca1:
			for letter2 in lca1:
				possibilities += 1
				answer = letter1 + letter2
				if x == answer:
					print("The answer is " + answer + ". It took " + "{:,}".format(possibilities) + " guesses.")
					return

	if(len(x) == 3):
		for letter1 in lca1:
			for letter2 in lca1:
				for letter3 in lca1:
					possibilities += 1
					answer = letter1 + letter2 + letter3
					if x == answer:
						print("The answer is " + answer + ". It took " + "{:,}".format(possibilities) + " guesses.")
						return
	if(len(x) == 4):
		for letter1 in lca1:
			for letter2 in lca1:
				for letter3 in lca1:
					for letter4 in lca1:
						possibilities += 1
						answer = letter1 + letter2 + letter3 + letter4
						if x == answer:
							print("The answer is " + answer + ". It took " + "{:,}".format(possibilities) + " guesses.")
							return

	if(len(x) == 5):
		for letter1 in lca1:
			for letter2 in lca1:
				for letter3 in lca1:
					for letter4 in lca1:
						for letter5 in lca1:
							possibilities += 1
							answer = letter1 + letter2 + letter3 + letter4 + letter5
							if x == answer:
								print("The answer is " + answer + ". It took " + "{:,}".format(possibilities) + " guesses.")
								return

	if(len(x) == 6):
		for letter1 in lca1:
			for letter2 in lca1:
				for letter3 in lca1:
					for letter4 in lca1:
						for letter5 in lca1:
							for letter6 in lca1:
								possibilities += 1
								answer = letter1 + letter2 + letter3 + letter4 + letter5 + letter6
								if x == answer:
									print("The answer is " + answer + ". It took " + "{:,}".format(possibilities) + " guesses.")
									return

	if(len(x) == 7):
		for letter1 in lca1:
			for letter2 in lca1:
				for letter3 in lca1:
					for letter4 in lca1:
						for letter5 in lca1:
							for letter6 in lca1:
								for letter7 in lca1:
									possibilities += 1
									answer = letter1 + letter2 + letter3 + letter4 + letter5 + letter6 + letter7
									if x == answer:
										print("The answer is " + answer + ". It took " + "{:,}".format(possibilities) + " guesses.")
										return

start = time.process_time()
brute()
print("It took " + str(time.process_time() - start) + " seconds.\n\n")

start = time.process_time()
brute()
print("It took " + str(time.process_time() - start) + " seconds.\n\n")

start = time.process_time()
brute()
print("It took " + str(time.process_time() - start) + " seconds.\n\n")

start = time.process_time()
brute()
print("It took " + str(time.process_time() - start) + " seconds.\n\n")

start = time.process_time()
brute()
print("It took " + str(time.process_time() - start) + " seconds.\n\n")

start = time.process_time()
brute()
print("It took " + str(time.process_time() - start) + " seconds.\n\n")

start = time.process_time()
brute()
print("It took " + str(time.process_time() - start) + " seconds.\n\n")
```

### This program was run on 7 different inputs. The amount of guesses made are recorded along with the time it took to arrive at the correct password. We can see from the output that guessing a 7 character password took this algorithm roughly 12.8 minutes. Increasing the amount of characters, even by just 1, will increase the time taken by a very substantial amount. Passwords that exceed length 13, even if they consist of only lowercase characters, take a very large amount of time to bruteforce.<br><br>
